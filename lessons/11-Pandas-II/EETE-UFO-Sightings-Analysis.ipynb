## Complete working code

```python
import pandas as pd
from IPython.display import display,HTML
from ipywidgets import widgets, interact_manual
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_colwidth', None)


def loadUFODataFrame():
    ufo_df = pd.DataFrame()
    for i in range(1,6):
        ufo = pd.read_csv(f'https://raw.githubusercontent.com/mafudge/datasets/master/ufo-sightings/ufo-sightings-2016-0{i}.csv')
        ufo_df = pd.concat([ufo_df, ufo], ignore_index=True)
    return ufo_df


df = loadUFODataFrame()
display(HTML("<h1>Search UFO Sightings<h1>"))
states = list(df['State'].dropna().unique())
states.sort()
states.insert(0,'*ANY*')
shapes = list(df['Shape'].dropna().unique())
shapes.sort()
shapes.insert(0, '*ANY*')
@interact_manual(state=states, shape=shapes, summary="")
def main(state,shape, summary):
    search_df = df
    if state != '*ANY*':
        search_df = search_df[ search_df.State == state ]
    if shape != '*ANY*':
        search_df = search_df[ search_df.Shape == shape ]
    if summary != "":    
        search_df = search_df[search_df.Summary.str.find(summary) >=0 ]
    display(search_df)
```
